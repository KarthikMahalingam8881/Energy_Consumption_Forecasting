## Data Cleaning

```python
merged_data.isnull().sum()
```

```python
merged_data.shape
```

```python
irrelevant_cols = ['hotwater', 'chilledwater', 'steam', 'water', 'irrigation', 'solar',
                   'gas', 'industry', 'subindustry', 'heatingtype', 'date_opened',
                   'numberoffloors', 'occupants', 'energystarscore', 'leed_level', 'rating',
                   'site_eui', 'source_eui']
merged_data=merged_data.drop(columns=irrelevant_cols)
```

```python
# Filling the missing weather datas using forward-fill and interpolation
merged_data['airTemperature'] = merged_data['airTemperature'].fillna(method='ffill')
merged_data['dewTemperature'] = merged_data['dewTemperature'].fillna(method='ffill')
merged_data['cloudCoverage'] = merged_data['cloudCoverage'].interpolate()
```

```python
merged_data = merged_data.drop(columns=['precipDepth6HR'])
```

```python
#Imputing energy_usage using the mean for each building and hour
#Filling using the mean energy usage for the same building across the same time of day
merged_data['energy_usage'] = merged_data.groupby(['building_id', merged_data['timestamp'].dt.hour])['energy_usage'].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
# Forward fill and interpolate for the missing weather data
weather_columns = ['precipDepth1HR', 'seaLvlPressure', 'windDirection', 'cloudCoverage']
for col in weather_columns:
    merged_data[col] = merged_data[col].fillna(method='ffill').interpolate()
```

```python
#Normalizing energy usage by sqft- to have a fair comparison for different buildings
merged_data['energy_per_sqft'] = merged_data['energy_usage'] / merged_data['sqft']
```

```python
merged_data = merged_data.drop(['site_id_x'], axis=1) # dropping site-id_x as it is similar to site_id_y
```

```python
merged_data = merged_data.drop(['site_id_kaggle'], axis=1)
```

```python
merged_data.columns = [col.lower().strip() for col in merged_data.columns]
```

```python
# removing extreme values in energy_usage
merged_data['energy_usage'] = merged_data['energy_usage'].clip(upper=merged_data['energy_usage'].quantile(0.99))
```

```python
merged_data['cloudcoverage'] = merged_data['cloudcoverage'].fillna(merged_data['cloudcoverage'].median())
merged_data['sealvlpressure'] = merged_data['sealvlpressure'].fillna(merged_data['sealvlpressure'].median())
```

## forecasting

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
import xgboost as xgb
import matplotlib.pyplot as plt

data=merged_data
#numerical_columns = ['energy_per_sqft', 'airtemperature', 'dewtemperature', 'windspeed', 'cloudcoverage','yearbuilt','precipdepth1hr','sealvlpressure']
columns = ['building_id', 'energy_per_sqft', 'airtemperature', 'dewtemperature', 'windspeed', 'cloudcoverage','yearbuilt','precipdepth1hr','sealvlpressure']
data = data[columns]
# data = data[numerical_columns]

def create_lagged_data(df, target_col, lags=3):
    for lag in range(1, lags + 1):
        df[f'{target_col}_lag{lag}'] = df[target_col].shift(lag)
    return df

def normalize_building_data(building_data, scaler):
    columns_to_normalize = ['energy_per_sqft', 'airtemperature', 'dewtemperature', 'windspeed', 'cloudcoverage']
    building_data[columns_to_normalize] = scaler.fit_transform(building_data[columns_to_normalize])
    return building_data

scaler = MinMaxScaler()
```

```python
# for each building_id
results = {}
for building in data['building_id'].unique():
    print(f"Processing Building: {building}")

    building_data = data[data['building_id'] == building].copy()   # for the specific building

    building_data = create_lagged_data(building_data, 'energy_per_sqft', lags=3)
    building_data.dropna(inplace=True)

    # Droping non-numeric(building_id) columns
    building_data = building_data.drop(columns=['building_id'])

    # Normalize data
    building_data = normalize_building_data(building_data, scaler)

    # Train/Test split
    train_size = int(0.8 * len(building_data))
    train_data = building_data[:train_size]
    test_data = building_data[train_size:]

    X_train = train_data.drop(columns=['energy_per_sqft'])
    y_train = train_data['energy_per_sqft']
    X_test = test_data.drop(columns=['energy_per_sqft'])
    y_test = test_data['energy_per_sqft']

    # Reshape data for LSTM (samples, timesteps, features)
    X_train_lstm = X_train.values.reshape((X_train.shape[0], 1, X_train.shape[1]))
    X_test_lstm = X_test.values.reshape((X_test.shape[0], 1, X_test.shape[1]))

    # Initialize models
    lstm_model = Sequential()
    lstm_model.add(LSTM(50, activation='relu', input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])))
    lstm_model.add(Dense(1))
    lstm_model.compile(optimizer='adam', loss='mse')

    lstm_model.fit(X_train_lstm, y_train, epochs=10, batch_size=32, verbose=0)# Training- LSTM

    y_pred_lstm = lstm_model.predict(X_test_lstm) # Prediction - LSTM

    xgb_model = xgb.XGBRegressor(objective='reg:squarederror', n_estimators=100, learning_rate=0.1)
    xgb_model.fit(X_train, y_train)    # Training- XGBoost
    y_pred_xgb = xgb_model.predict(X_test)    # Prediction-XGBoost

    # Calculating RMSE for bot the models
    rmse_lstm = np.sqrt(mean_squared_error(y_test, y_pred_lstm))
    rmse_xgb = np.sqrt(mean_squared_error(y_test, y_pred_xgb))

    # results
    results[building] = {
        'rmse_lstm': rmse_lstm,
        'rmse_xgb': rmse_xgb,
        'y_test': y_test,
        'y_pred_lstm': y_pred_lstm.flatten(),
        'y_pred_xgb': y_pred_xgb
    }

    print(f"Building: {building}, LSTM RMSE: {rmse_lstm:.4f}, XGBoost RMSE: {rmse_xgb:.4f}")
```

```python
# Aggregate RMSE for all buildings
rmse_summary = pd.DataFrame([
    {'Building': building, 'LSTM RMSE': results[building]['rmse_lstm'],
     'XGBoost RMSE': results[building]['rmse_xgb']}
    for building in results
])

# RMSE comparison
print("### RMSE Summary for All Buildings ###")
print(rmse_summary)

rmse_summary.set_index('Building').plot(kind='bar', figsize=(12, 6), title="Model RMSE Comparison")
plt.ylabel("RMSE")
plt.show()
```

```python
# all predictions from XGBoost
forecasted_energy = []
actual_energy = []

for building in results:
    forecasted_energy.extend(results[building]['y_pred_xgb'])
    actual_energy.extend(results[building]['y_test'])

forecast_df = pd.DataFrame({
    'Actual': actual_energy,
    'Predicted': forecasted_energy
})

# Actual vs Predicted
plt.figure(figsize=(12, 6))
plt.plot(forecast_df['Actual'], label="Actual Energy per sqft", color='black')
plt.plot(forecast_df['Predicted'], label="XGBoost Predicted Energy per sqft", color='red')
plt.title("XGBoost: Actual vs Predicted Energy per sqft (Aggregated)")
plt.xlabel("Time Points")
plt.ylabel("Energy per sqft (Normalized)")
plt.legend()
plt.grid()
plt.show()
```

```python
sample_building = list(results.keys())[0]
y_test = results[sample_building]['y_test']
y_pred_lstm = results[sample_building]['y_pred_lstm']
y_pred_xgb = results[sample_building]['y_pred_xgb']

plt.figure(figsize=(12, 6))
plt.plot(y_test.index, y_test, label='Actual', color='black')
plt.plot(y_test.index, y_pred_lstm, label='LSTM Prediction', color='blue')
plt.plot(y_test.index, y_pred_xgb, label='XGBoost Prediction', color='red')
plt.title(f"Predictions for Building: {sample_building}")
plt.xlabel("Time")
plt.ylabel("Energy Per Sqft (Normalized)")
plt.legend()
plt.show()
```

### Forecast Future Energy Costs

```python
building_costs = {}

# Looping through each building to calculate predicted costs
for building in results:
    predicted_energy = results[building]['y_pred_xgb']
    predicted_cost = predicted_energy * cost_per_kwh

    # the total predicted cost
    total_cost = np.sum(predicted_cost)
    building_costs[building] = total_cost

cost_df = pd.DataFrame.from_dict(building_costs, orient='index', columns=['Projected_Cost ($)'])
cost_df = cost_df.sort_values(by='Projected_Cost ($)', ascending=False)

# projected energy costs per building
print("### Projected Future Energy Costs Per Building ###")
print(cost_df)

plt.figure(figsize=(12, 6))
cost_df.plot(kind='bar', legend=False, color='red')
plt.title("Projected Future Energy Costs Per Building")
plt.ylabel("Projected Cost ($)")
plt.xlabel("Building ID")
plt.tight_layout()
plt.show()
```

```python
# Calculate total energy usage per building
total_energy_usage = merged_data.groupby('building_id')['energy_usage'].sum()

# Combining efficiency, total energy usage, and projected costs
efficiency_cost_df = pd.DataFrame({
    'Average_Energy_Per_Sqft': merged_data.groupby('building_id')['energy_per_sqft'].mean(),
    'Total_Energy_Usage': total_energy_usage,
    'Projected_Cost': cost_df['Projected_Cost ($)']
})

efficiency_cost_df = efficiency_cost_df.sort_values(by='Projected_Cost', ascending=False)

print("### Combined Analysis: Efficiency, Usage, and Costs ###")
# print(efficiency_cost_df)

plt.figure(figsize=(14, 4))
plt.scatter(efficiency_cost_df['Total_Energy_Usage'], efficiency_cost_df['Average_Energy_Per_Sqft'],
            s=efficiency_cost_df['Projected_Cost'] / 5, c='red', alpha=0.5)
plt.title("Energy Efficiency vs Total Usage (Bubble Size = Projected Cost)")
plt.xlabel("Total Energy Usage (kWh)")
plt.ylabel("Average Energy per Sqft (kWh/sqft)")
plt.grid()
plt.show()
```
